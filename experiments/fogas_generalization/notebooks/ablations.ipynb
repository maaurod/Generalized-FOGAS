Deterministic grid:

```python
from types import SimpleNamespace
import torch

det_states = torch.arange(25, dtype=torch.long)
det_actions = torch.arange(4, dtype=torch.long)

det_N = len(det_states)
det_A = len(det_actions)
det_gamma = 0.99
det_x_0 = 0

det_grid_size = 5
det_goal_grid = 24
det_pit_grid = 18

det_wall_states = {6, 7, 12}
det_terminal_states = {det_goal_grid, det_pit_grid}


def det_state_to_pos(s):
    return divmod(int(s), det_grid_size)


def det_pos_to_state(row, col):
    return row * det_grid_size + col


def det_move_deterministic(s, a):
    s = int(s)
    a = int(a)

    if s in det_terminal_states:
        return s

    row, col = det_state_to_pos(s)

    if a == 0:
        new_row, new_col = row - 1, col
    elif a == 1:
        new_row, new_col = row + 1, col
    elif a == 2:
        new_row, new_col = row, col - 1
    elif a == 3:
        new_row, new_col = row, col + 1
    else:
        raise ValueError("action must be in {0, 1, 2, 3}")

    if not (0 <= new_row < det_grid_size and 0 <= new_col < det_grid_size):
        return s

    sp = det_pos_to_state(new_row, new_col)

    if sp in det_wall_states:
        return s

    return sp


def det_next_state(s, a):
    return det_move_deterministic(s, a)


def det_transition_probs(s, a, intended_prob=None):
    return {det_next_state(s, a): 1.0}


def det_transition_fn(s, a):
    probs = torch.zeros(det_N, dtype=torch.float64)
    probs[det_next_state(s, a)] = 1.0
    return probs


def det_reward_fn(s, a):
    sp = det_next_state(s, a)

    if sp == det_goal_grid:
        return 1.0

    if sp == det_pit_grid:
        return -1.0

    return -0.01


det_mdp = DiscreteMDP(
    states=det_states,
    actions=det_actions,
    gamma=det_gamma,
    x0=det_x_0,
    reward_fn=det_reward_fn,
    transition_fn=det_transition_fn,
    terminal_states=list(det_terminal_states),
).to(DEVICE)

det_planner = Planner(det_mdp)

det_grid = SimpleNamespace(
    name="Deterministic",
    states=det_states,
    actions=det_actions,
    N=det_N,
    A=det_A,
    gamma=det_gamma,
    x_0=det_x_0,
    grid_size=det_grid_size,
    goal_grid=det_goal_grid,
    pit_grid=det_pit_grid,
    wall_states=det_wall_states,
    terminal_states=det_terminal_states,
    state_to_pos=det_state_to_pos,
    transition_probs=det_transition_probs,
    mdp=det_mdp,
    planner=det_planner,
)
```

Stochastic grid

```python
stoch_states = torch.arange(25, dtype=torch.long)
stoch_actions = torch.arange(4, dtype=torch.long)

stoch_N = len(stoch_states)
stoch_A = len(stoch_actions)
stoch_gamma = 0.9
stoch_x_0 = 0

dataset_path_5grid_stoch = str(DATASETS_DIR / "5grid_stochastic.csv")

stoch_grid_size = 5
stoch_goal_grid = 24
stoch_pit_grid = 18

stoch_wall_states = {6, 7, 12}
stoch_terminal_states = {stoch_goal_grid, stoch_pit_grid}


def stoch_state_to_pos(s):
    return divmod(int(s), stoch_grid_size)


def stoch_pos_to_state(row, col):
    return row * stoch_grid_size + col


def stoch_move_deterministic(s, a):
    s = int(s)
    a = int(a)

    if s in stoch_terminal_states:
        return s

    row, col = stoch_state_to_pos(s)

    if a == 0:
        new_row, new_col = row - 1, col
    elif a == 1:
        new_row, new_col = row + 1, col
    elif a == 2:
        new_row, new_col = row, col - 1
    elif a == 3:
        new_row, new_col = row, col + 1
    else:
        raise ValueError("action must be in {0, 1, 2, 3}")

    if not (0 <= new_row < stoch_grid_size and 0 <= new_col < stoch_grid_size):
        return s

    sp = stoch_pos_to_state(new_row, new_col)

    if sp in stoch_wall_states:
        return s

    return sp


def stoch_transition_probs(s, a, intended_prob=0.8):
    s = int(s)
    a = int(a)

    probs_by_state = {}

    for candidate_a in range(stoch_A):
        prob = (1.0 - intended_prob) / stoch_A
        if candidate_a == a:
            prob += intended_prob

        sp = stoch_move_deterministic(s, candidate_a)
        probs_by_state[sp] = probs_by_state.get(sp, 0.0) + prob

    return probs_by_state


def stoch_transition_fn(s, a):
    probs = torch.zeros(stoch_N, dtype=torch.float64)

    for sp, prob in stoch_transition_probs(s, a).items():
        probs[sp] = prob

    return probs


def stoch_next_state(s, a):
    probs = stoch_transition_fn(s, a)
    return int(torch.multinomial(probs.float(), num_samples=1).item())


def stoch_reward_from_next_state(sp):
    sp = int(sp)

    if sp == stoch_goal_grid:
        return 1.0
    if sp == stoch_pit_grid:
        return -1.0

    return -0.1


def stoch_reward_fn(s, a):
    return sum(
        prob * stoch_reward_from_next_state(sp)
        for sp, prob in stoch_transition_probs(s, a).items()
    )


stoch_mdp = DiscreteMDP(
    states=stoch_states,
    actions=stoch_actions,
    gamma=stoch_gamma,
    x0=stoch_x_0,
    reward_fn=stoch_reward_fn,
    transition_fn=stoch_transition_fn,
    terminal_states=list(stoch_terminal_states),
).to(DEVICE)

stoch_planner = Planner(stoch_mdp)

stoch_grid = SimpleNamespace(
    name="Stochastic",
    states=stoch_states,
    actions=stoch_actions,
    N=stoch_N,
    A=stoch_A,
    gamma=stoch_gamma,
    x_0=stoch_x_0,
    grid_size=stoch_grid_size,
    goal_grid=stoch_goal_grid,
    pit_grid=stoch_pit_grid,
    wall_states=stoch_wall_states,
    terminal_states=stoch_terminal_states,
    state_to_pos=stoch_state_to_pos,
    transition_probs=stoch_transition_probs,
    mdp=stoch_mdp,
    planner=stoch_planner,
)
```

Plotting

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import Rectangle, FancyArrowPatch, Circle


def _to_numpy(x):
    if x is None:
        return None
    if hasattr(x, "detach"):
        x = x.detach().cpu().numpy()
    return np.asarray(x)


def _draw_gridworld(
    ax,
    env,
    values=None,
    policy=None,
    policy_mode="probs",
    agent_state=None,
    transition_state=None,
    transition_action=None,
    intended_prob=0.8,
    title=None,
    show_state_ids=True,
):
    values = _to_numpy(values)
    if values is not None:
        values = values.reshape(-1)

    policy = _to_numpy(policy)

    action_vec = {
        0: (0.0, -0.34),
        1: (0.0, 0.34),
        2: (-0.34, 0.0),
        3: (0.34, 0.0),
    }

    facecolors = {
        "empty": "#F8F9FA",
        "wall": "#2B2D30",
        "start": "#4C78A8",
        "goal": "#2CA02C",
        "pit": "#D62728",
    }

    def state_kind(s):
        if s in env.wall_states:
            return "wall"
        if s == env.x_0:
            return "start"
        if s == env.goal_grid:
            return "goal"
        if env.pit_grid is not None and s == env.pit_grid:
            return "pit"
        return "empty"

    ax.set_xlim(-0.5, env.grid_size - 0.5)
    ax.set_ylim(env.grid_size - 0.5, -0.5)
    ax.set_aspect("equal")

    for s in range(env.N):
        r, c = env.state_to_pos(s)
        kind = state_kind(s)

        ax.add_patch(
            Rectangle(
                (c - 0.5, r - 0.5),
                1,
                1,
                facecolor=facecolors[kind],
                edgecolor="none",
                linewidth=0,
                zorder=1,
            )
        )

        if show_state_ids and kind == "empty":
            ax.text(c - 0.42, r - 0.36, str(s), fontsize=7, color="#666666", zorder=4)

        if kind in {"wall", "start", "goal", "pit"}:
            label = {"wall": "W", "start": "S", "goal": "G", "pit": "P"}[kind]
            ax.text(c, r, label, fontsize=15, fontweight="bold", color="white",
                    ha="center", va="center", zorder=5)

        if values is not None and kind != "wall":
            ax.text(c, r + 0.34, f"{values[s]:.2f}", fontsize=8,
                    ha="center", va="center", zorder=6)

    ax.set_xticks(np.arange(env.grid_size))
    ax.set_yticks(np.arange(env.grid_size))
    ax.set_xticks(np.arange(-0.5, env.grid_size, 1), minor=True)
    ax.set_yticks(np.arange(-0.5, env.grid_size, 1), minor=True)
    ax.grid(which="minor", color="#555555", linewidth=0.8)
    ax.tick_params(which="minor", bottom=False, left=False)

    if policy is not None:
        for s in range(env.N):
            if s in env.wall_states or s in env.terminal_states:
                continue

            r, c = env.state_to_pos(s)

            if policy.ndim == 1 or policy_mode == "best":
                a = int(policy[s]) if policy.ndim == 1 else int(np.argmax(policy[s]))
                actions_to_draw = [(a, 1.0)]
            else:
                row = np.asarray(policy[s], dtype=float)
                if row.sum() <= 0 or not np.isclose(row.sum(), 1.0, atol=1e-3):
                    row = row - row.min()
                    row = row / row.sum() if row.sum() > 0 else np.ones(env.A) / env.A
                actions_to_draw = [(a, row[a]) for a in range(env.A) if row[a] > 1e-3]

            for a, p in actions_to_draw:
                dx, dy = action_vec[int(a)]

                ax.add_patch(
                    FancyArrowPatch(
                        (c, r),
                        (c + dx, r + dy),
                        arrowstyle="-|>",
                        mutation_scale=10 + 8 * float(p),
                        linewidth=0.7 + 2.6 * float(p),
                        color="black",
                        alpha=0.25 + 0.75 * float(p),
                        zorder=8,
                    )
                )

                if policy.ndim == 2 and policy_mode == "probs":
                    ax.text(c + dx * 1.15, r + dy * 1.15, f"{p:.2f}",
                            fontsize=6.5, ha="center", va="center", zorder=9)

    if transition_state is not None and transition_action is not None:
        s = int(transition_state)
        a = int(transition_action)
        r, c = env.state_to_pos(s)

        ax.add_patch(
            Rectangle(
                (c - 0.48, r - 0.48),
                0.96,
                0.96,
                fill=False,
                edgecolor="#FFB000",
                linewidth=2.5,
                zorder=12,
            )
        )

        probs = env.transition_probs(s, a, intended_prob=intended_prob)

        for sp, prob in probs.items():
            rr, cc = env.state_to_pos(sp)

            if sp == s:
                continue

            ax.add_patch(
                FancyArrowPatch(
                    (c, r),
                    (cc, rr),
                    arrowstyle="-|>",
                    mutation_scale=14,
                    linewidth=1.4 + 3.0 * prob,
                    color="#FFB000",
                    alpha=0.9,
                    connectionstyle="arc3,rad=0.12",
                    zorder=11,
                )
            )

            ax.text(
                (c + cc) / 2,
                (r + rr) / 2,
                f"{prob:.2f}",
                fontsize=8,
                fontweight="bold",
                color="#7A4B00",
                ha="center",
                va="center",
                bbox=dict(facecolor="white", edgecolor="none", alpha=0.85, pad=1.0),
                zorder=14,
            )

    if agent_state is not None:
        r, c = env.state_to_pos(agent_state)
        ax.add_patch(
            Circle((c, r), radius=0.16, facecolor="#FFD166",
                   edgecolor="black", linewidth=1.2, zorder=20)
        )

    ax.set_xlabel("Column", fontsize=10)
    ax.set_ylabel("Row", fontsize=10)
    ax.tick_params(labelsize=9)
    ax.set_title(title or env.name, fontsize=13, pad=10)


def plot_5grid_pair(
    det_env,
    stoch_env,
    det_values=None,
    stoch_values=None,
    det_policy=None,
    stoch_policy=None,
    det_transition_state=13,
    det_transition_action=3,
    stoch_transition_state=13,
    stoch_transition_action=3,
    intended_prob=0.8,
    save_path=None,
    dpi=300,
):
    fig, axes = plt.subplots(1, 2, figsize=(11.5, 5.6))

    _draw_gridworld(
        axes[0],
        det_env,
        values=det_values,
        policy=det_policy,
        policy_mode="best",
        agent_state=det_transition_state,
        transition_state=det_transition_state,
        transition_action=det_transition_action,
        intended_prob=intended_prob,
        title="Deterministic Grid World",
    )

    _draw_gridworld(
        axes[1],
        stoch_env,
        values=stoch_values,
        policy=stoch_policy,
        policy_mode="probs",
        agent_state=stoch_transition_state,
        transition_state=stoch_transition_state,
        transition_action=stoch_transition_action,
        intended_prob=intended_prob,
        title="Stochastic Grid World",
    )

    legend_handles = [
        mpatches.Patch(color="#F8F9FA", label="Empty"),
        mpatches.Patch(color="#2B2D30", label="Wall"),
        mpatches.Patch(color="#4C78A8", label="Start"),
        mpatches.Patch(color="#2CA02C", label="Goal"),
        mpatches.Patch(color="#D62728", label="Pit"),
        mpatches.Patch(color="#FFB000", label="Selected action & transition probabilities"),
        mpatches.Patch(color="#FFD166", label="Agent"),
    ]

    fig.legend(
        handles=legend_handles,
        loc="lower center",
        ncol=7,
        frameon=False,
        fontsize=9,
        bbox_to_anchor=(0.5, -0.02),
    )

    plt.tight_layout(rect=[0, 0.07, 1, 1])

    if save_path is not None:
        fig.savefig(save_path, dpi=dpi, bbox_inches="tight")

    plt.show()
```

```python
plot_5grid_pair(
    det_grid,
    stoch_grid,
    det_transition_state=13,
    det_transition_action=3,
    stoch_transition_state=13,
    stoch_transition_action=3,
    intended_prob=0.8,
)
```

# Beta ablations

```python
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

try:
    from IPython.display import display
except ImportError:
    display = print


def find_repo_root(start=None):
    start = Path.cwd() if start is None else Path(start)
    for p in [start, *start.parents]:
        if (p / "src" / "rl_methods").exists() and (p / "data").exists():
            return p
    return Path.cwd()


ROOT = find_repo_root()
RESULTS_DIR = ROOT / "data" / "results" / "generalization"
BETA_CSV = RESULTS_DIR / "ablations" / "beta" / "final_linear_5grid_tabular_beta_ablation.csv"

METRICS = [
    "greedy_success_rate",
    "solver_success_rate",
    "greedy_avg_reward",
    "solver_avg_reward",
]

PROBLEM_ORDER = ["deterministic", "stochastic"]
T_ORDER = [1500, 3000]

UPDATE_ORDER = [
    "fogas_full",
    "fogas_diag",
    "metric_no_stabilization",
    "euclidean_stabilized",
    "fenchel_mirror",
    "fenchel_br",
    "projected_gradient",
]

RANK_COLUMNS = [
    "greedy_success_rate",
    "solver_success_rate",
    "greedy_avg_reward",
    "solver_avg_reward",
    "elapsed_seconds",
]

RANK_ASCENDING = [
    False,
    False,
    False,
    False,
    True,
]

if not BETA_CSV.exists():
    raise FileNotFoundError(f"Missing beta ablation CSV: {BETA_CSV}")

df = pd.read_csv(BETA_CSV)

if "status" in df.columns:
    df = df[df["status"].eq("ok")].copy()

if df.empty:
    raise ValueError("Beta ablation CSV has no successful rows.")

rename = {
    "solver_avg_return": "solver_avg_reward",
    "greedy_avg_return": "greedy_avg_reward",
}
df = df.rename(columns={k: v for k, v in rename.items() if k in df.columns})

missing = [c for c in ["problem", "beta_update", "T", "eta", "rho", *METRICS] if c not in df.columns]
if missing:
    raise ValueError(f"Missing required columns: {missing}")

df["run_order"] = df["run_idx"] if "run_idx" in df.columns else np.arange(len(df))

df["problem"] = pd.Categorical(df["problem"], categories=PROBLEM_ORDER, ordered=True)
df["beta_update"] = pd.Categorical(df["beta_update"], categories=UPDATE_ORDER, ordered=True)

rank_cols = [c for c in RANK_COLUMNS if c in df.columns]
rank_ascending = RANK_ASCENDING[: len(rank_cols)]


def best_rows_per_group(data, group_cols):
    return (
        data.sort_values(group_cols + rank_cols, ascending=[True] * len(group_cols) + rank_ascending)
        .groupby(group_cols, as_index=False, observed=True)
        .head(1)
        .reset_index(drop=True)
    )


metric_titles = {
    "greedy_success_rate": "Greedy success rate",
    "solver_success_rate": "Solver success rate",
    "greedy_avg_reward": "Greedy average reward",
    "solver_avg_reward": "Solver average reward",
}


# ============================================================
# 1. Original plots: best row of each beta update
# ============================================================

plot_df = df[df["T"].isin(T_ORDER)].copy()

best_by_problem_t_update = best_rows_per_group(
    plot_df,
    ["problem", "T", "beta_update"],
)

for problem in PROBLEM_ORDER:
    for T in T_ORDER:
        sub = best_by_problem_t_update[
            best_by_problem_t_update["problem"].astype(str).eq(problem)
            & best_by_problem_t_update["T"].eq(T)
        ].copy()

        if sub.empty:
            print(f"No rows for problem={problem}, T={T}")
            continue

        sub = sub.sort_values(rank_cols, ascending=rank_ascending)
        updates = sub["beta_update"].astype(str).tolist()

        fig, axes = plt.subplots(2, 2, figsize=(15, 9))
        axes = axes.ravel()

        for ax, metric in zip(axes, METRICS):
            values = sub[metric].astype(float).values

            colors = [
                "tab:blue" if update == "fogas_full"
                else "tab:orange" if update == "fogas_diag"
                else "tab:gray"
                for update in updates
            ]

            ax.bar(updates, values, color=colors, alpha=0.85)
            ax.set_title(metric_titles.get(metric, metric))
            ax.set_ylabel(metric)
            ax.grid(True, axis="y", alpha=0.25)
            ax.tick_params(axis="x", rotation=35)

            for idx, value in enumerate(values):
                ax.text(idx, value, f"{value:.3g}", ha="center", va="bottom", fontsize=9)

        fig.suptitle(
            f"{problem.capitalize()} - best result per beta update, T={T}",
            fontsize=15,
        )
        fig.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()


# ============================================================
# 2. fogas_full sweep plots: rho and eta
# ============================================================

fogas_full_df = df[df["beta_update"].astype(str).eq("fogas_full")].copy()

if "sweep_family" in fogas_full_df.columns:
    sweep_df = fogas_full_df[fogas_full_df["sweep_family"].astype(str).eq("eta_rho_sweep")].copy()
    if sweep_df.empty:
        sweep_df = fogas_full_df.copy()
else:
    sweep_df = fogas_full_df.copy()


def plot_fogas_full_sweep(data, sweep_param):
    if sweep_param not in data.columns:
        print(f"No column found for sweep parameter: {sweep_param}")
        return

    for problem in PROBLEM_ORDER:
        for T in T_ORDER:
            sub = data[
                data["problem"].astype(str).eq(problem)
                & data["T"].eq(T)
                & data[sweep_param].notna()
            ].copy()

            if sub.empty:
                print(f"No fogas_full rows for problem={problem}, T={T}, sweep={sweep_param}")
                continue

            fixed_param = "eta" if sweep_param == "rho" else "rho"

            if fixed_param in sub.columns:
                best_fixed = best_rows_per_group(sub, [fixed_param]).iloc[0][fixed_param]
                sub = sub[sub[fixed_param].eq(best_fixed)].copy()
                fixed_label = f", fixed {fixed_param}={best_fixed:g}"
            else:
                fixed_label = ""

            sub = (
                sub.sort_values([sweep_param, *rank_cols], ascending=[True] + rank_ascending)
                .groupby(sweep_param, as_index=False, observed=True)
                .head(1)
                .sort_values(sweep_param)
            )

            fig, axes = plt.subplots(2, 2, figsize=(15, 9))
            axes = axes.ravel()

            x = sub[sweep_param].astype(float).values

            for ax, metric in zip(axes, METRICS):
                y = sub[metric].astype(float).values

                ax.plot(x, y, marker="o", linewidth=2, color="tab:blue")
                ax.set_title(metric_titles.get(metric, metric))
                ax.set_xlabel(sweep_param)
                ax.set_ylabel(metric)
                ax.grid(True, alpha=0.25)

                if sweep_param in {"rho", "eta"}:
                    ax.set_xscale("log")

                for xi, yi in zip(x, y):
                    ax.text(xi, yi, f"{yi:.3g}", ha="center", va="bottom", fontsize=9)

            fig.suptitle(
                f"{problem.capitalize()} - fogas_full {sweep_param} sweep, T={T}{fixed_label}",
                fontsize=15,
            )
            fig.tight_layout(rect=[0, 0, 1, 0.95])
            plt.show()
```

```python
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def find_repo_root(start=None):
    start = Path.cwd() if start is None else Path(start)
    for p in [start, *start.parents]:
        if (p / "src" / "rl_methods").exists() and (p / "data").exists():
            return p
    return Path.cwd()


ROOT = find_repo_root()

THETA_CURVES_CSV = (
    ROOT
    / "data"
    / "results"
    / "generalization"
    / "ablations"
    / "theta"
    / "final_linear_5grid_tabular_theta_ablation_curves.csv"
)

if not THETA_CURVES_CSV.exists():
    matches = list(
        (ROOT / "data" / "results" / "generalization").rglob(
            "final_linear_5grid_tabular_theta_ablation_curves.csv"
        )
    )
    if not matches:
        raise FileNotFoundError(f"Could not find theta ablation curves CSV under {ROOT}")
    THETA_CURVES_CSV = matches[0]

raw = pd.read_csv(THETA_CURVES_CSV)

raw = raw[raw["checkpoint_step"].eq(raw["checkpoint_step"].max())].copy()
raw = raw[raw["problem"].eq("stochastic")].copy()

group_cols = [
    "problem",
    "ablation",
    "theta_mode",
    "theta_start_mode",
    "theta_lambda",
    "D_theta",
    "theta_lr",
    "theta_inner_steps",
    "rho",
    "T",
    "checkpoint_step",
]

curves = (
    raw.groupby(group_cols, dropna=False, as_index=False)
    .agg(
        seed_count=("seed", "nunique"),
        solver_avg_reward_mean=("solver_avg_reward", "mean"),
        solver_avg_reward_std=("solver_avg_reward", "std"),
    )
)

curves["solver_avg_reward_std"] = curves["solver_avg_reward_std"].fillna(0.0)

plt.rcParams.update({
    "figure.dpi": 150,
    "savefig.dpi": 300,
    "font.size": 10,
    "axes.titlesize": 11,
    "axes.labelsize": 10,
    "legend.fontsize": 8.5,
    "xtick.labelsize": 9,
    "ytick.labelsize": 9,
    "axes.spines.top": False,
    "axes.spines.right": False,
})

start_colors = {
    "warm": "#1f77b4",
    "zero": "#d62728",
}

start_labels = {
    "warm": "Warm start",
    "zero": "Zero start",
}

start_markers = {
    "warm": "o",
    "zero": "s",
}


def plot_theta_sweep(df, sweep_col, title, xlabel, xscale="log", base=None):
    fig, ax = plt.subplots(figsize=(6.2, 4.2))

    for start in ["warm", "zero"]:
        sub = df[df["theta_start_mode"].eq(start)].sort_values(sweep_col)

        if sub.empty:
            continue

        x = sub[sweep_col].to_numpy(dtype=float)
        y = sub["solver_avg_reward_mean"].to_numpy(dtype=float)
        std = sub["solver_avg_reward_std"].fillna(0.0).to_numpy(dtype=float)

        ax.fill_between(
            x,
            y - std,
            y + std,
            color=start_colors[start],
            alpha=0.25,
            linewidth=0,
            zorder=1,
        )

        ax.plot(
            x,
            y,
            color=start_colors[start],
            marker=start_markers[start],
            markersize=4.5,
            linewidth=2.0,
            label=start_labels[start],
            zorder=2,
        )

    if xscale == "log" and base is not None:
        ax.set_xscale("log", base=base)
    elif xscale:
        ax.set_xscale(xscale)

    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel("Average episode reward")

    ax.set_ylim(-2.5, 0.5)
    ax.set_yticks(np.arange(-2.5, 0.51, 0.5))

    ax.grid(True, axis="y", alpha=0.25)
    ax.grid(True, axis="x", alpha=0.10)

    ax.legend(
        title="Theta initialization",
        loc="upper left",
        frameon=True,
        fontsize=8.5,
        title_fontsize=9,
    )

    fig.tight_layout()
    plt.show()


lambda_sweep = curves[curves["ablation"].eq("reg_fixed_lambda")].copy()
inner_sweep = curves[curves["ablation"].eq("reg_fixed_inner_steps")].copy()

plot_theta_sweep(
    lambda_sweep,
    sweep_col="theta_lambda",
    title="Stochastic 5-grid: theta regularization ablation",
    xlabel=r"$\lambda_\theta$",
    xscale="log",
)

plot_theta_sweep(
    inner_sweep,
    sweep_col="theta_inner_steps",
    title="Stochastic 5-grid: theta inner-steps ablation",
    xlabel=r"$K_Q$",
    xscale="log",
    base=2,
)
```
